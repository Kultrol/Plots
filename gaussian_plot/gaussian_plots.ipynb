```python
import numpy as np 
import matplotlib.pyplot as plt 

lambda_parameters = [0.5, 1.0, 5.0, 10.0, 15.0] #Controls Width
a = 0 #Center Point

for lam in lambda_parameters:
    x = np.linspace(-10,10,1000)
    y = np.sqrt(lam / np.pi) * np.exp(-lam * (x-a)**2)
    plt.plot(x,y, label=f'Width Param={lam}')

plt.grid(True)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Gaussian')
plt.legend()

plt.xlim(-3.5,3.5)
plt.ylim(-1,4)

plt.show()
```

```python
import numpy as np 
import plotly.express as px 

lam = 5 #Controls Width
a = 0 #Center Point

x = np.linspace(-3.5,3.5,1000)
y = np.sqrt(lam / np.pi) * np.exp(-lam * (x-a)**2)

fig = px.line(x=x,y=y, title='Guassian Function')
fig.show()
```

```python
import numpy as np 
import plotly.graph_objects as go
fig = go.Figure()


lambda_parameters = [0.5, 1.0, 5.0, 10.0, 15.0] #Controls Width
a = 0 #Center Point

for lam in lambda_parameters:
    x = np.linspace(-3.5,3.5,1000)
    y = np.sqrt(lam / np.pi) * np.exp(-lam * (x-a)**2)
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name=f'lambda={lam}'))

fig.update_layout(title='Multiple Gaussians', xaxis_title='x', yaxis_title='p(x)')
fig.show()
```
